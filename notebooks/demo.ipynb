```python
# Pierwsze wywołanie: Numba kompiluje JIT
t0 = time.perf_counter()
sum_of_squares(1_000_000)
t1 = time.perf_counter()
print(f'Pierwsze wywołanie (z kompilacją JIT): {t1 - t0:.3f}s')

# Kolejne wywołanie: już skompilowane, dużo szybsze
t0 = time.perf_counter()
result = sum_of_squares(1_000_000)
t1 = time.perf_counter()
print(f'Kolejne wywołanie (skompilowane):      {t1 - t0:.6f}s')
print(f'Wynik: {result:.0f}')
```

```python
# Porównanie z czystym Pythonem
def sum_of_squares_python(n):
    total = 0.0
    for i in range(n):
        total += i * i
    return total

N = 1_000_000

t0 = time.perf_counter()
sum_of_squares_python(N)
t1 = time.perf_counter()
python_time = t1 - t0

t0 = time.perf_counter()
sum_of_squares(N)
t1 = time.perf_counter()
numba_time = t1 - t0

print(f'Python: {python_time:.4f}s')
print(f'Numba:  {numba_time:.6f}s')
print(f'Przyspieszenie: {python_time / numba_time:.0f}x')
```
